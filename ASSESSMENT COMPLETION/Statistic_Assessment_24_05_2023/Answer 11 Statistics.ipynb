```python
import scipy.stats as stats
import statsmodels.api as sm

# Assuming you have the blood pressure data stored in variables 'data' and 'classes'

# a. One-way ANOVA test to check the equality of distributions
f_value, p_value = stats.f_oneway(data[' Blood Pressure Before (mmHg)'], data[' Blood Pressure After (mmHg)'])
if p_value < 0.05:
    print("The distribution of classes is not the same.")
else:
    print("The distribution of classes is the same.")

# b. Levene's test for equality of variances
w_value, p_value = stats.levene(data[' Blood Pressure Before (mmHg)'], data[' Blood Pressure After (mmHg)'])
if p_value < 0.05:
    print("The variances are not equal across classes.")
else:
    print("The variances are equal across classes.")

# c. Classification using LDA and QDA
# insufficient data

# d. Pairwise t-tests for equality of means
pairwise_ttests = sm.stats.multicomp.pairwise_tukeyhsd(data[' Blood Pressure Before (mmHg)'], data[' Blood Pressure After (mmHg)'])
print(pairwise_ttests.summary())
```

```python
# d. Check the equality of means for all the classes:
# Perform one-way ANOVA
f_value, p_value = stats.f_oneway(data[' Blood Pressure Before (mmHg)'], data[' Blood Pressure After (mmHg)'])

if p_value > 0.05:
    print("Means are equal")
else:
    print("Means are not equal")
```
