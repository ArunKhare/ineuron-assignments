#### Question 1: 
 - INPUT STRING, COUNT FREQUENCY OF EACH WORD, 
        THEIR MIGHT BE REPEATED CHARACTERS IN THE STRING.
    FIND HIGHEST FREQUENCY AND RETURNS THE LENGTH OF HIGHEST -FREQUENCY WORD

```python
sentence = "She wore a beautiful dress, dress, dress, dress, dress, dress, dress, dress, and danced danced danced danced danced danced danced danced all night long."

def max_freq_and_length(sentence):

    list_words = sentence.split()

    word_frequencies= {}
    for word in list_words:
        if word not in word_frequencies:
            word_frequencies[word] =1
        else:
            word_frequencies[word] +=1

    max_val = 0
    
    for key, val in word_frequencies.items():
        if val > max_val:
            max_val=val
    
    max_keys = []
    word_length = []
    
    for key, val in word_frequencies.items():
        if word_frequencies[key] == max_val:
            word_length.append(len(key))
            max_keys.append(key)
            
    print(f"Words and freq: {word_frequencies}")
    print(f"max_freq ====================: {max_val}")
    print(f"words with highest frequencies:{max_keys}" )
    print(f"length of words===============:{word_length}\n")

max_freq_and_length(sentence)
```

```python
import random

def generate_sentence():
    sentence = ""
    words = ["apple", "banana", "orange", "grape", "kiwi"]
    for _ in range(random.randint(2, 5)):
        word = random.choice(words)
        repetition = random.randint(2, 6)
        second_word = random.choice(words)
        second_repetition = random.randint(2,6)
        sentence += f"{word} " * repetition + f"{second_word} " * second_repetition
    return sentence.strip()

sentences = [generate_sentence() for _ in range(5)]
print(sentences)
for sentence in sentences:
    max_freq_and_length(sentence)
```

#### Question 3:
    -Write a program, which would download the data from the provided link, and then read the data and convert
        that into properly structured data and return it in Excel format.

```python
Link = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"

import requests
import pandas as pd 
import json
import openpyxl

r = requests.get(Link,)
pokemon_data   = r.json()


pokemon_json = json.dumps(pokemon_data["pokemon"], indent=4)  # Convert Python object to JSON string
df = pd.read_json(path_or_buf=pokemon_json,orient="records") # convert Json string to pd.DataFrame

# ExcelWriter : Class for writing DataFrame objects into excel sheets.
# Engine to use for writing. 
excel_writer = pd.ExcelWriter("pokemon.xlsx",engine="openpyxl") 

# Write object to an Excel sheet.
xls = df.to_excel(excel_writer=excel_writer, index=False)
excel_writer.close()
```
