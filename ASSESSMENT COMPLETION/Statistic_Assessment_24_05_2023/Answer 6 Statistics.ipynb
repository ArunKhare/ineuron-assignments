```python
import pandas as pd
import numpy as np
import matplotlib_inline
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\arunk\Assignments\ASSESSMENT COMPLETION\Statistic_Assessment_24_05_2023\data.csv")
shape=data.shape
```

```python
from scipy import stats
# The Shapiro-Wilk test is used to determine if a dataset follows a normal distribution.
def determine_distriution(feature):
    _, p_value_before = stats.shapiro(feature)
    _, p_value_after = stats.shapiro(feature)

    print(p_value_before ,p_value_after)
    # Interpret the results
    alpha = 0.05
    if p_value_before > alpha:
        print("it suggests that there is not enough evidence to reject the null hypothesis. The null hypothesis assumes that the data is normally distributed.\n")
    else:
        print(" it suggests that there is sufficient evidence to reject the null hypothesis. It indicates that the data may not follow a normal distribution.\n")

print(" Distribution of Blood Pressure Before")
determine_distriution( data[" Blood Pressure Before (mmHg)"])

print('Distribution of Blood Pressure After')
determine_distriution(data[' Blood Pressure After (mmHg)'])
```

```python
def plot_distribution(feature):
    plt.figure(figsize=(8, 6))
    plt.hist(feature, bins=10, alpha=0.5)
    plt.xlabel('Data')
    plt.ylabel('Freqency')
    plt.title('Histogram')
    plt.show()
    stats.probplot(data[" Blood Pressure Before (mmHg)"], plot=plt)
    plt.xlabel('Theoretical Quantiles')
    plt.ylabel('Ordered Values')
    plt.title('Q-Q Plot')
    plt.show()

plot_distribution(data[" Blood Pressure Before (mmHg)"])
```

```python
plot_distribution(data[" Blood Pressure After (mmHg)"])
```

Here are a few key observations:

Curvature or bending away from the line: points show curvature or bending away from the line, it indicates deviations from the assumed distribution. This could imply skewness or heavy-tailedness in the dataset.

Deviations at the tails: points deviate from the line at the tails, it suggests differences in the tails' behavior. This could indicate heavier or lighter tails compared to the assumed distribution.
